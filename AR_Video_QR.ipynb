```python
import qrcode

url = 'https://raw.githubusercontent.com/Sillerus/AR-Video-Try/refs/heads/main/IMG_5706.mp4'

qr = qrcode.QRCode(
    version=1,  # Размер QR (1-40, auto по умолчанию)
    error_correction=qrcode.constants.ERROR_CORRECT_L,  # Уровень коррекции ошибок (L=7%, M=15%, Q=25%, H=30%)
    box_size=10,  # Размер пикселя
    border=4,  # Отступы
)
qr.add_data(url)
qr.make(fit=True)

# Создание и сохранение изображения
img = qr.make_image(fill_color="black", back_color="white")
img.save('qr_code.png')
print("QR-код сохранён как qr_code.png")
img  # Отобразит QR в Colab
```
